```python
f = open("cs621_text_data_1.txt")
```

```python
# process text
data_raw = []
f.seek(0)
for line in f:
  line = line.strip()
  line = line.replace('.','')
  line = line.replace(',','')
  data_raw.append(line)
f.close()
```

```python
print(data_raw)
```

```python
# split into list of words
data = []
for i in range(len(data_raw)):
  words = data_raw[i].split()
  for w in words:
    w = w.lower()
    data.append(w)
```

```python
print(data)
```

```python
data_distinct = list(set(data))
```

```python
print(data_distinct)
```

```python
word_freq = {}
for w in data_distinct:
  word_freq[w] = 0
```

```python
for i in range(len(data_distinct)):
  key_word = data_distinct[i]
  count = 0
  for w in data:
    if w == key_word:
      count = count + 1
    word_freq[key_word] = count
```

```python
print(word_freq)
```

```python
word_freq = {}
for w in data:
  if w not in word_freq:
    word_freq[w] = 1
  else:
    word_freq[w] +=1
```

```python
print(word_freq)
```

```python
import pandas as pd
word_freq_series = pd.Series(word_freq)
word_freq_series.sort_index(inplace=True)
```

```python
print(word_freq_series)
```
